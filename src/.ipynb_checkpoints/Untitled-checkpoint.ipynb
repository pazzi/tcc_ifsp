```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("../data/final.csv", sep=";")
```

```python
df['valor']
```

```python
df['valor'] = df['valor'].astype(float)
```

```python
df3=df[df['class_id']==3] #Florest
```

```python
df3
```

```python
df3['valor'] = df['valor'] * 0.01
```

```python
plt.axis(ymin=0, ymax=28)
plt.plot(df3['ano'], df3['valor'])
plt.grid()
```

```python
df20=df[df['class_id']==20]  #cana de a√ßucar
```

```python
df20['valor'] = df['valor'] * 0.01
```

```python
plt.axis(ymin=0, ymax=200)
plt.plot(df20['ano'], df20['valor'])
plt.grid()
```

```python
df24=df[df['class_id']==24]  #urbano
df24['valor'] = df['valor'] * 0.01
```

```python
plt.axis(ymin=0, ymax=25)
plt.plot(df24['ano'], df24['valor'])
plt.grid()
```

```python
plt.axis(ymin=0, ymax=200)
plt.plot(df24['ano'], df24['valor'], label='Urbana')
plt.plot(df20['ano'], df20['valor'], label='Cana de a√ßucar')
plt.plot(df3['ano'], df3['valor'], label='Floresta')
plt.legend()
plt.grid()
```
